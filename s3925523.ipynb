## Data Pre- Processing

### **Explanation of Data Preprocessing Steps and Justification**

#### **1. Train-Test Splitting**
**Why?**
- Splitting the dataset into **training (80%)** and **testing (20%)** sets ensures that the model is evaluated on unseen data.
- This prevents **data leakage** and provides a more accurate assessment of model generalization.

**Steps Taken:**
- Used `train_test_split()` from `sklearn.model_selection` to randomly split the data.
- Maintained **random_state=42** for reproducibility.
- Both **`TARGET_Capacity` and `TARGET_CaseCount`** were split separately.


#### **2. Feature Scaling**
**Why?**
- Different features have varying ranges, which can affect model performance.
- Some models (e.g., **Linear Regression, k-NN, Neural Networks**) require **scaled data** to perform optimally.

##### **2.1 Min-Max Scaling (Normalization)**
**Why?**
- Transforms all feature values into a **0 to 1 range**.
- Helps models that rely on distance-based calculations (e.g., **k-NN, Neural Networks**).

**Steps Taken:**
- Applied `MinMaxScaler()` from `sklearn.preprocessing`.
- **Fit on training data** and **transformed both training and test data**.

##### **2.2 Standardization (Z-score Normalization)**
**Why?**
- Converts data to have **mean = 0** and **standard deviation = 1**.
- Useful for models like **Linear Regression, Poisson Regression**, which assume normally distributed inputs.

**Steps Taken:**
- Applied `StandardScaler()` from `sklearn.preprocessing`.
- **Fit on training data** and **transformed both training and test data**.

---

### **Why No Other Preprocessing Was Performed?**
- **No missing values** were detected, so **imputation was not necessary**.
- **No categorical features** in the dataset, so **encoding was not needed**.
- **Feature selection was not performed** at this stage, as baseline models should first be tested with all features.

---

### **Final Justification**
âœ” **Train-Test Split** ensures unbiased evaluation.  
âœ” **Min-Max Scaling** helps distance-based models.  
âœ” **Standardization** improves regression-based models.  
âœ” **No unnecessary preprocessing** was applied, avoiding overcomplication.

```python
## TODO
## You need to run this code, so the output is displayed below

## 1. Train-Test Splitting

from sklearn.model_selection import train_test_split
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Load dataset
train_data = pd.read_csv("Data.csv")

# Separate features (X) and target variables (y)
X = train_data.drop(columns=["TARGET_Capacity", "TARGET_CaseCount"])
y_capacity = train_data["TARGET_Capacity"]
y_casecount = train_data["TARGET_CaseCount"]

# Split dataset into training (80%) and testing (20%) sets using scikit-learn's train_test_split()
X_train, X_test, y_train_capacity, y_test_capacity = train_test_split(
    X, y_capacity, test_size=0.2, random_state=42, shuffle=True
)

X_train, X_test, y_train_casecount, y_test_casecount = train_test_split(
    X, y_casecount, test_size=0.2, random_state=42, shuffle=True
)

# Display dataset split details
print(f"Original Dataset Size: {train_data.shape[0]}")
print(f"Training Set Size: {X_train.shape[0]}")
print(f"Testing Set Size: {X_test.shape[0]}")

## 2. Feature Scaling
## 2.1 Min-Max Scaling (0-1 Normalization)

# Apply Min-Max Scaling
minmax_scaler = MinMaxScaler()
X_train_minmax = minmax_scaler.fit_transform(X_train)
X_test_minmax = minmax_scaler.transform(X_test)

# Convert back to DataFrame
X_train_minmax_df = pd.DataFrame(X_train_minmax, columns=X_train.columns)

# Display summary statistics after scaling
print("Min-Max Scaled Training Data:")
display(X_train_minmax_df.describe())

## 2.2 Standardization (Z-score Normalization)

# Apply Standardization
standard_scaler = StandardScaler()
X_train_standard = standard_scaler.fit_transform(X_train)
X_test_standard = standard_scaler.transform(X_test)

# Convert back to DataFrame
X_train_standard_df = pd.DataFrame(X_train_standard, columns=X_train.columns)

# Display summary statistics after standardization
print("Standardized Training Data:")
display(X_train_standard_df.describe())

## 3. Visualizing the Effect of Scaling Using Histograms

# Select a feature to demonstrate scaling effects
feature = "ComputeUtilizationRate"

plt.figure(figsize=(15, 5))

# Original Feature Distribution
plt.subplot(1, 3, 1)
sns.histplot(train_data[feature], bins=30, color='blue', alpha=0.6, kde=True)
plt.title(f"Original Distribution: {feature}")

# Min-Max Scaled Distribution
plt.subplot(1, 3, 2)
sns.histplot(X_train_minmax_df[feature], bins=30, color='red', alpha=0.6, kde=True)
plt.title(f"Min-Max Scaled: {feature}")

# Standardized Distribution
plt.subplot(1, 3, 3)
sns.histplot(X_train_standard_df[feature], bins=30, color='green', alpha=0.6, kde=True)
plt.title(f"Standardized: {feature}")

plt.tight_layout()
plt.show()
```

## Why Use Linear Regression for `TARGET_Capacity`?
Linear regression is a fundamental model used to predict continuous numerical values. Since `TARGET_Capacity` represents the chatbot's conversation length (a continuous variable), linear regression serves as a good baseline.

### **Advantages:**
- Simple and interpretable.
- Efficient for small to medium-sized datasets.
- Provides a benchmark before trying more complex models.

### **Potential Issues:**
- Assumes a **linear** relationship between features and target.
- Sensitive to outliers.
- Can underperform if the data has a **non-linear** pattern.

If performance is suboptimal, we might try **Polynomial Regression or GLM** to capture more complex patterns.


## Why Use Poisson Regression for `TARGET_CaseCount`?
Since `TARGET_CaseCount` represents the **number of tasks handled in parallel** (a count variable), Poisson regression is more appropriate.

### **Advantages:**
- Designed for count data (non-negative integers).
- Accounts for skewed distributions (common in count data).
- Ensures that predictions remain non-negative.

### **Limitations:**
- Assumes **mean = variance** (which might not hold).
- May underperform if overdispersion is present (variance > mean).

If overdispersion is detected, a **Negative Binomial Regression** could be considered

Performance metrics help assess how well the model predicts outcomes. Since we are dealing with **regression problems**, the following metrics are relevant:

## **1. Mean Squared Error (MSE)**
- Measures the average squared difference between predicted and actual values.
- **Lower MSE = Better model performance.**
- Sensitive to **outliers** due to squaring errors.

## **2. Mean Absolute Error (MAE)**
- Similar to MSE, but less sensitive to outliers.
- Calculates the absolute difference instead of squaring.

## **3. R-Squared (RÂ²)**
- Measures the proportion of variance explained by the model.
- Ranges from **0 to 1** (higher = better).
- Used primarily for `TARGET_Capacity` (since itâ€™s continuous).

## **4. Deviance for Poisson Regression**
- Used to assess goodness of fit for count data models.
- Lower deviance indicates better model fit.

Parameter tuning optimizes the model to balance **complexity and performance**.

#### **1. Hyperparameter Tuning for Linear Regression**
- The standard linear regression model has **no hyperparameters** to tune.
- However, **Ridge Regression (L2 Regularization)** can be used to prevent overfitting.

#### Balancing Model Complexity with Performance
- **Underfitting** (High Bias):
  - Model is too simple (e.g., using only Linear Regression).
  - Poor training and test performance.
  
- **Overfitting** (High Variance):
  - Model is too complex (e.g., overly high-degree polynomial regression).
  - Good training performance but poor test generalization.

### **Best Practices:**

```python
## TODO
## You need to run this code, so the output is displayed bellow
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, PoissonRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score

# LINEAR REGRESSION FOR TARGET_CAPACITY
def linear_regression(X_train, X_test, y_train, y_test):
    model = LinearRegression()
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    
    mse = mean_squared_error(y_test, predictions)
    r2 = r2_score(y_test, predictions)
    
    print("Parameters of the Linear model:", model.coef_)
    print("Intercept of the Linear model:", model.intercept_)
    print("Linear Regression - MSE:", mse)
    print("Linear Regression - RÂ² Score:", r2)
    
    # Visualizing parameters
    plt.figure(figsize=(10, 5))
    plt.bar(range(len(model.coef_)), model.coef_)
    plt.xlabel("Feature Index")
    plt.ylabel("Coefficient Value")
    plt.title("Linear Model Coefficients for TARGET_Capacity")
    plt.show()
    
    return model, mse, r2

# POISSON REGRESSION FOR TARGET_CASECOUNT
def poisson_regression(X_train, X_test, y_train, y_test):
    model = PoissonRegressor(alpha=0.01, max_iter=500)
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    
    mse = mean_squared_error(y_test, predictions)
    r2 = r2_score(y_test, predictions)
    
    print("Poisson Regression - MSE:", mse)
    print("Poisson Regression - RÂ² Score:", r2)
    
    return model, mse, r2



# EXECUTE MODELS AND OUTPUT RESULTS
lin_reg_results = linear_regression(X_train_standard, X_test_standard, y_train_capacity, y_test_capacity)
poisson_reg_results = poisson_regression(X_train_standard, X_test_standard, y_train_casecount, y_test_casecount)

# Compare models
results = pd.DataFrame({
    'Model': ['Linear Regression (TARGET_Capacity)', 'Poisson Regression (TARGET_CaseCount)'],
    'MSE': [lin_reg_results[1], poisson_reg_results[1]],
    'R2 Score': [lin_reg_results[2], poisson_reg_results[2]]
})

from IPython.display import display

display(results)
```

```python
## TODO
## You need to run this code, so the output is displayed bellow
# Polynomial Regression for TARGET_Capacity
def polynomial_regression(X_train, X_test, y_train, y_test, degree=2):
    poly = PolynomialFeatures(degree)
    X_train_poly = poly.fit_transform(X_train)
    X_test_poly = poly.transform(X_test)
    
    model = LinearRegression()
    model.fit(X_train_poly, y_train)
    predictions = model.predict(X_test_poly)
    return model, mean_squared_error(y_test, predictions), r2_score(y_test, predictions)

# GLM for TARGET_Capacity
def glm_capacity(X_train, X_test, y_train, y_test):
    model = sm.OLS(y_train, sm.add_constant(X_train)).fit()
    predictions = model.predict(sm.add_constant(X_test))
    return model, mean_squared_error(y_test, predictions), r2_score(y_test, predictions)

# GLM for TARGET_CaseCount (Poisson family)
def glm_casecount(X_train, X_test, y_train, y_test):
    model = sm.GLM(y_train, sm.add_constant(X_train), family=sm.families.Poisson()).fit()
    predictions = model.predict(sm.add_constant(X_test))
    return model, mean_squared_error(y_test, predictions), r2_score(y_test, predictions)

# PERFORMANCE METRICS SELECTION

# Execute models
poly_reg_results = polynomial_regression(X_train_standard, X_test_standard, y_train_capacity, y_test_capacity)
glm_c_results = glm_capacity(X_train_standard, X_test_standard, y_train_capacity, y_test_capacity)
glm_p_results = glm_casecount(X_train_standard, X_test_standard, y_train_casecount, y_test_casecount)

# Compare models
results = pd.DataFrame({
    'Model': ['Linear Regression', 'Poisson Regression', 'Polynomial Regression', 'GLM (OLS)', 'GLM (Poisson)'],
    'MSE': [lin_reg_results[1], poisson_reg_results[1], poly_reg_results[1], glm_c_results[1], glm_p_results[1]],
    'R2 Score': [lin_reg_results[2], poisson_reg_results[2], poly_reg_results[2], glm_c_results[2], glm_p_results[2]]
})

print(results)
```
